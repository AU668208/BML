## Sammenligning af ADC, T2 og DWI
I koden herunder sammenlignes de 3 typer MRI billeder fra datasættet.
Det ses at alle 3D-billederne har samme dimensioner 270x270x24.


Herunder vises 3 slices af billederne fra de 3 typer billeder. Det ses at der er en større visuel sammenhæng mellem ADC og T2. DWI skiller sig lidt ud.

I studiet hvor de bruger datasættet laver de både anatomisk segmentering for at finde anatomiske 'landmarks'. Til dette bruges T2 som input T2_anatomy_reader1 som label.
De laver også tumor segmentering og her bruges både T2, ADC og DWI som input og T2_tumor_reader1 som label.

```python
path = 'C:/Users/kvjkv/Downloads/prostate158_train/prostate158_train/'
n_patient = '063'
```

```python
ADCpath = path + 'train/' + n_patient + '/adc.nii.gz'
ADCimg = nib.load(ADCpath).get_fdata()
T2path = path +'train/' + n_patient + '/T2.nii.gz'
T2img = nib.load(T2path).get_fdata()
DWIpath = path + 'train/' + n_patient + '/dwi.nii.gz'
DWIimg = nib.load(DWIpath).get_fdata()
print(ADCimg.shape)
print(T2img.shape)
print(DWIimg.shape)

frame = 5

plt.subplot(1,3,1)
plt.imshow(ADCimg[:,:,frame])
plt.subplot(1,3,2)
plt.imshow(T2img[:,:,frame])
plt.subplot(1,3,3)
plt.imshow(DWIimg[:,:,frame])
```
